####l) Determine qual atributo apresenta maior variabilidade dentre os atributos: anos de educação, anos de experiência e idade.

```
# coeficiente de variação = desvio padrao/media

dp_education_yrs = df.education_yrs.std(axis= 0)
dp_experience_yrs = df.experience_yrs.std(axis = 0)
dp_age = df.age.std(axis = 0)

md_education_yrs = df.education_yrs.mean(axis= 0)
md_experience_yrs = df.experience_yrs.mean(axis = 0)
md_age = df.age.mean(axis = 0)


cv_edy = (dp_education_yrs/md_education_yrs)*100
cv_exy = (dp_experience_yrs/md_experience_yrs)*100
cv_a = (dp_age/md_age)*100

atributos = [cv_edy, cv_exy, cv_a]
var = max(atributos, key=int)

print('Desv. Padrão Anos de educação: %.2f' % dp_education_yrs)
print('Desv. Padrão Anos de experiência: %.2f' % dp_experience_yrs)
print('Desv. Padrão Idade: %.2f' % dp_age)

print('\nMédia Anos de educação: %.2f' % md_education_yrs)
print('Média Anos de experiência: %.2f' % md_experience_yrs)
print('Média Idade: %.2f' % md_age)

print('\nCoef. Var. Anos de educação: %.2f' % cv_edy)
print('Coef. Var. Anos de experiência: %.2f' % cv_exy)
print('Coef. Var. Idade: %.2f' % cv_a)

print('\nMaior Variabilidade =  %.2f' % var)
```
