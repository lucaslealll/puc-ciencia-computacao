###1-) SUPONHA QUE A IDADE MÉDIA DE APOSENTADORIA PARA TODA A POPULAÇÃO DE UM PAÍS É DE 64 ANOS E A DISTRIBUIÇÃO É **NORMAL** COM UM DESVIO PADRÃO DE 3 ANOS.

```python
med = 64 # Média
dsv_pdr = 3 # Desv. padrão

#68%
i68 = med - dsv_pdr # = 61
f68 = med + dsv_pdr # = 67
area68 = 34

#95%
i95 = med - 2*dsv_pdr # = 58
f95 = med + 2*dsv_pdr # = 70
area95 = 13.5

#99.7%
i99 = med - 3*dsv_pdr # = 55
f99 = med + 3*dsv_pdr # = 73
area99 = 2.35
```

###a-) Qual e a faixa etária aproximada em que 68% das pessoas se aposentam? 

```python
print("{} à {} ".format(i68, f68))
```

###b-) Qual e a faixa etária aproximada em que 95% das pessoas se aposentam? 

```python
print("{} à {} ".format(i95, f95))
```

###c-) Qual o percentual aproximado da população que se aposenta entre 55 e 67 anos?

```python
print("{} à {} ".format(i99, f68))
area = area99 + area95 + area68 + area68 # 2.35% + 13.5% + 34% + 34%
print("{:.2f}%".format(area))
```

###2-) EXERCÍCIO Nº 14 DA UNIDADE 4 (SLIDE 77).
#### O número médio de acidentes mensais em um determinado cruzamento é três. Qual é a probabilidade de que em um determinado mês ocorram quatro acidentes no cruzamento?



```python
u = 3 # media
k = 4 # K

fx = stats.poisson.pmf(k, u)
fx = fx * 100

print("{:.2f}%".format(fx)) #aproximação de 2 casas decimais
```

###3-) EXERCÍCIO Nº 18 DA UNIDADE 4 (SLIDE 78).
#### Se a probabilidade de atingir um alvo num único disparo é 0,3, qual é a probabilidade de que em 4 disparos o alvo seja atingido no mínimo 3 vezes? 

```python
n = 4 # 4 disparos
x = 3 # acertar 3 vezes
p = 0.3 # prob. de acerto

fx = stats.binom.pmf(x, n, p)
fx = fx * 100

print("{:.2f}%".format(fx)) #aproximação de 2 casas decimais
```

###4-) EXERCÍCIO Nº 22 DA UNIDADE 4 (SLIDE 110).
#### O número de horas semanais que os adultos norte-americanos gastam em frente ao computador em casa é normalmente distribuído, com média de 5 horas e desvio padrão de 1 hora.  Um adulto foi selecionado ao acaso.


###a-) Obtenha a probabilidade de o número de horas gastas no computador pelo adulto ser inferior a 2,5 semanais.

```python
med = 5 # média (horas)
dsv_pdr = 1 # desvio padrão (horas)
```

```python
x1 = 2.5

ra = stats.norm.cdf(x1, med, dsv_pdr) * 100

print("{:.2f}%".format(ra)) #aproximação de 2 casas decimais
```

###b-) Obtenha a probabilidade de o número de horas gastas no computador pelo adulto estar entre 2,5 e 7,5.

```python
x2 = 7.5
rb = 0.62 #questao anterior

r2 = stats.norm.cdf(x2, med, dsv_pdr) * 100

print("{:.2f}%".format(r2-rb)) #aproximação de 2 casas decimais
```
