```python
import math
import numpy as np
from scipy import stats
from scipy.stats import norm, zscore
```

**DE UMA AMOSTRA ALEATÓRIA DE 50 DIAS DESTE ANO, OS PREÇOS DOS COMBUSTÍVEIS NOS EUA TIVERAM UMA MÉDIA DE `US$4,28`. SUPONHA QUE O DESVIO PADRÃO POPULACIONAL É DE `US$0,21`. DETERMINE O INTERVALO DE CONFIANÇA (ESTIMATIVA INTERVALAR) PARA A MÉDIA POPULACIONAL DO PREÇO DOS COMBUSTÍVEIS:**

1. **COM 95% DE CONFIANÇA;**
2. **COM 99% DE CONFIANÇA;**

```python
n = 50 # Amostra
med = 4.28
desv_pdr = 0.21
nconf1 = 0.95 # Nível de confiança 1
nconf2 = 0.99 # Nível de confiança 2
```

1. **COM 95% DE CONFIANÇA;**

```python
# Magem de erro E = (Nivel_Conf.Desv_Pdr) / raiz_amostra
# Intervalo =  med-E & med+E
intervalo = norm.interval(nconf1, med, desv_pdr/math.sqrt(n))
print('%.4f' % intervalo[0])
print('%.4f' % intervalo[1])
```

2. **COM 99% DE CONFIANÇA;**

```python
# Magem de erro E = (Nivel_Conf.Desv_Pdr) / raiz_amostra
# Intervalo =  med-E & med+E
intervalo = norm.interval(nconf2, med, desv_pdr/math.sqrt(n))
print('%.4f' % intervalo[0])
print('%.4f' % intervalo[1])
```

## **Forma manual**

```python
# Magem de erro E = (Nivel_Conf.Desv_Pdr) / raiz_amostra
margem = 1.96 * (desv_pdr)/math.sqrt(n)
print('Margem 95% = ' + str(margem))
print('Intervalo A = %.4f' % (med - margem))
print('Intervalo B = %.4f' % (med + margem))
```
