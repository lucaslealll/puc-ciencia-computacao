##Gráficos de linha


---



```python
import numpy as np
x = np.linspace(-5, 5)
y = x ** 2
plt.plot(x, y, 'r') # 'r' é a cor vermelha (red)
plt.xlabel('Eixo x')
plt.ylabel('Eixo x')
plt.title('y = x^2')
plt.show()
```

####Subplots (vários plots na mesma tela)

```python
plt.subplot(2,2,1) # 2x2
plt.plot(x, y, 'r--') 
plt.subplot(2,2,2)
plt.plot(y, x, 'g*-');
plt.subplot(2,2,3)
plt.plot(x, y, 'b+-');
plt.subplot(2,2,4)
plt.plot(y, x, 'k^-');
```

Outra maneira de plotar vários gráficos (subplots):

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# Aparência padrão 
axes[0].plot(x, y)
axes[0].grid(True)

# Aparência customizada 
axes[1].plot(x, y, x, x**3, lw=4)
axes[1].grid(color='k', alpha=0.5, linestyle='dashed', lw=2)
```

Salvando a figura em arquivo:

```python
fig.savefig("graflines.png")
```

### Gráfico de "enxame"
Outra forma de visualizar os mesmos dados do exemplo anterior é o "gráfico de enxame". Em vez de caixas (boxs), ele plota cada ponto de dados individualmente, mas os agrupa com base em sua distribuição.

```python
#ax=sns.swarmplot(x='Mfr Name', y='CombMPG', data=df)
ax=sns.stripplot(x='Mfr Name', y='CombMPG', data=df)
sns.set(rc={'figure.figsize':(20,5)})
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()
```

###Countplot (gráfico de contagem = "histograma")

Outra ferramenta é o "gráfico de contagem". É basicamente a mesma coisa que um histograma, mas para dados categóricos. Ele permite que você conte quantas vezes cada categoria no eixo X ocorre em seus dados e plote-os. 

```python
ax=sns.countplot(x='Mfr Name', data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45)
plt.show()
```

###Gráficos de superfícies

Para usar gráficos 3D no matplotlib, basta criar uma instância da classe Axes3D. Os eixos 3D podem ser adicionados a uma tela matplotlib exatamente da mesma maneira que os eixos 2D; ou passando um argumento de palavra-chave projection = '3d' para os métodos add_axes ou add_subplot.

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D

alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T

fig = plt.figure(figsize=(14,6))

# ax é uma instância de eixo com reconhecimento 3D 
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# surface_plot com classificação de cores e barra de cores
ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

#### Superfície de contorno com projeções nos planos

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1,1,1, projection='3d')

ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
cset = ax.contour(X, Y, Z, zdir='z', offset=-np.pi, cmap=plt.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-np.pi, cmap=plt.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=3*np.pi, cmap=plt.cm.coolwarm)

ax.set_xlim3d(-np.pi, 2*np.pi);
ax.set_ylim3d(0, 3*np.pi);
ax.set_zlim3d(-np.pi, 2*np.pi);
plt.show
```
